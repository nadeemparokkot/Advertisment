```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/advertising_data")
df
```

```python
#input and output
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x
```

```python
#to plot graph using seaborn(no matplotlib) 
#TV sales plotting
import seaborn as sns
sns.regplot(x=df['TV'],y=y,color='g')
```

```python
#radio sales plotting
sns.regplot(x=df['Radio'],y=y)
```

```python
#newspaper sales plottting
sns.regplot(x=df['Newspaper'],y=y,color='r')
```

```python
#train and test
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30)
ytest
```

```python
#model creation regression
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
ypred
```

```python
print(model.predict([[110.4,86.5,87.6]]))
```

```python
#y=m1x1+m2x2+.....+mnxn+c
print('constant',model.intercept_)
print('Slop is',model.coef_)
print('TV slop is',model.coef_[0])
print('radio slop is',model.coef_[1])
print('newspaper slop is',model.coef_[2])
list(zip(x,model.coef_))#to show 3 combine
```

```python
#create a data frame Actual value and predicted value
df1=pd.DataFrame({"Actual_value":ytest,'Predicted_value':ypred})
df1
```

```python
#perfomance evaluation(stight line(regression))
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print("MAE",mean_absolute_error(ytest,ypred))
print("MSE",mean_squared_error(ytest,ypred))
rmse=np.sqrt(mean_squared_error(ytest,ypred))
print("RMSE",rmse)
print("r2_score",r2_score(ytest,ypred))
```

```python
#model error
from sklearn.metrics import mean_absolute_percentage_error
print("percentage",mean_absolute_percentage_error(ytest,ypred)) 
```
